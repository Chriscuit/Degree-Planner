```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.ece.utexas.edu/undergraduate/courses')
soup = BeautifulSoup(page.content, 'html.parser')
data = soup.find_all('div', class_='view-content')
links = data[0].find_all('a')
for link in links:
    print(link['href'])
```

```python
for link in links:
    course = link['href'].split('/')[-1]
    #print(course)
    page = requests.get(link['href'])
    soup = BeautifulSoup(page.content, 'html.parser')
    div = soup.find_all('div', class_='panel-pane pane-entity-field pane-node-field-prerequisites no-title block')
    if div:
        div = div[0].find_all('div', class_='field-item even')
        text = div[0].find('p').text
        print(text, '\n')
```
