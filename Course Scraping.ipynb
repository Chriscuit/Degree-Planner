```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.ece.utexas.edu/undergraduate/courses')
soup = BeautifulSoup(page.content, 'html.parser')
data = soup.find_all('div', class_='view-content')
links = data[0].find_all('a')
for link in links:
    print(link['href'])
```
